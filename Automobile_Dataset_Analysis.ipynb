```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plot
sns.set(style="whitegrid")

# Set the figure size
plt.figure(figsize=(15, 7))

# Plotting categorical counts for 'body-style'
plt.subplot(2, 2, 1)
sns.countplot(x='body-style', data=df, palette='pastel')
plt.title('Distribution of Body Styles')

# Plotting numerical distribution for 'length'
plt.subplot(2, 2, 2)
sns.histplot(df['length'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Car Lengths')

# Plotting categorical counts for 'engine'
plt.subplot(2, 2, 3)
sns.countplot(x='engine', data=df, palette='muted')
plt.title('Distribution of Engine Types')

# Plotting numerical distribution for 'mileage'
plt.subplot(2, 2, 4)
sns.histplot(df['mileage'], bins=15, kde=True, color='salmon')
plt.title('Distribution of Mileage')

# Adjust layout for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

### We can see that cars with lesser length are giving more mileage and cars with more length are giving lesser mileage and the highest mileage is given by ohc engine


## Now let's make a regression or correlation plot to understand the correlation between length and mileage
### With the plot it is clear that their is a negative correlation between length and mileage which implies smaller(less) length cars have more mileage and vice a versa


```python
# Plotting scatter plot between Length and Mileage with engine as the hue parameter (to understand the variation)
sns.scatterplot(df, x=df["length"], y=df["mileage"], hue=df["engine"])



# Adding a regression line (correlation line)
sns.regplot(x='length', y='mileage', data=df, scatter=False, color='gray')

# Calculate and display the correlation coefficient
correlation_coefficient = df["length"].corr(df["mileage"])
plt.text(0.5, 0.95, f'Correlation: {correlation_coefficient:.2f}', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=12)

# Adding Labels
plt.title("Length and Mileage of Automobiles with engine")
plt.xlabel("Length of the Automobile (cm)")
plt.ylabel("Mileage (MPG)")
# Show the plot
plt.show()
```

For car buyers, this data is crucial, especially if fuel efficiency matters. Smaller cars excel for city driving with easy maneuverability and potential fuel savings. Conversely, those prioritizing space, comfort, or specific needs may accept lower mileage for other features in larger vehicles.

## Plotting Box Plot according to body-style and mean of length and mileage

```python
## Plotting Box Plot according to body-style and  mileage

# Grouping data by 'company' and 'body style', then calculating the mean of 'length' and 'mileage'
grouped_data = df.groupby(['company', 'body-style']).mean().reset_index()

sns.boxplot(x='body-style', y='mileage', data=grouped_data)
plt.title('Box Plot of Average Mileage by Body Style')
plt.xlabel('Body Style')
plt.ylabel('Average Mileage')
plt.show()
```

```python
# Grouping data by 'company' and 'body style', then calculating the mean of 'length' and 'mileage'
grouped_data = df.groupby(['company', 'body-style']).mean().reset_index()

## Plotting Box Plot according to body-style and mean of length and mileage
sns.boxplot(x='body-style', y='length', data=grouped_data)
plt.title('Box Plot of Average Length by Body Style')
plt.xlabel('Body Style')
plt.ylabel('Average Length (cm)')
plt.show()

```

**Summary:**

- **Hatchbacks:** Compact size, good mileage.
  
- **Sedans/Wagons:** Varied sizes, good mileage.
  
- **Convertibles:** Medium length, good mileage.
  
- **Hardtops:** Longer length, lesser mileage.

Consider body style for insights into both size and mileage when making a car purchase.

## Grouping the data by 'company' and 'body-style', then calculate the count

```python
# Grouping the data by 'company' and 'body-style', then calculate the count
grouped_data = df.groupby(['company', 'body-style']).size().reset_index(name='vehicle_count')
print("Dataset according to company and body style is:\n")
print(grouped_data)
```

```python
# Seeing which company produces the highest number of any body-style of vehicle
max_vehicle_count = grouped_data['vehicle_count'].max()
max_vehicle_company = grouped_data[grouped_data['vehicle_count'] == max_vehicle_count]

print("Company with the highest vehicle count for a body style:\n")
print(max_vehicle_company)
```

```python
# Grouping data by 'company' and calculating the mean of 'mileage'
average_mileage_by_company = df.groupby('company')['mileage'].mean().reset_index()

# Sorting the data by average mileage in descending order
average_mileage_by_company = average_mileage_by_company.sort_values(by='mileage', ascending=False)

# Set the figure size
plt.figure(figsize=(12, 8))

# Create a horizontal bar chart
sns.barplot(x='mileage', y='company', data=average_mileage_by_company, palette='muted')

# Adding Labels and increasing title size
plt.title('Average Mileage by Company', fontsize=20)
plt.xlabel('Average Mileage')
plt.ylabel('Company')

# Show the plot
plt.show()
```
